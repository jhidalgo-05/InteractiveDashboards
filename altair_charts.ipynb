```python
import zipfile
import pandas as pd

zip_path = r"C:\Users\joaqu\OneDrive\Desktop\Data Science Prog\HW3.zip"

with zipfile.ZipFile(zip_path) as z:
    with z.open("clean_football_data.csv") as f:
        df = pd.read_csv(f)
```

TASK

ALTAIR

```python
import altair as alt

scatter = alt.Chart(df).mark_point().encode(
    x='num_seasons:Q',
    y='total_seasons_country:Q'
).properties(
    title="Scatterplot of num_seasons vs total_seasons_country"
)
```

```python
bar = alt.Chart(df).mark_bar().encode(
    x='league_type:N',         # categorical field
    y='count():Q',             # aggregate count
    color='league_type:N'      # color by league_type
).properties(
    title="Number of records per league_type"
)

bar_text = bar.mark_text(
    dy=-5,              # shift text upward
    color='black'
).encode(
    text='count():Q'
)

bar_layered = alt.layer(bar, bar_text).properties(
    title="Number of records per league_type (Layered)"
)
```

TASK 3

```python
brush = alt.selection_interval(encodings=['x', 'y'])

scatter = alt.Chart(df).mark_point().encode(
    x='num_seasons:Q',
    y='total_seasons_country:Q',
    tooltip=[
        alt.Tooltip('num_seasons:Q', title='Num Seasons'),
        alt.Tooltip('total_seasons_country:Q', title='Total Seasons Country'),
        alt.Tooltip('league_type:N', title='League Type')
    ],
    color=alt.condition(brush, 'league_type:N', alt.value('lightgray'))  # highlight selected points
).add_selection(brush).properties(
    title="Scatterplot with Brush Selection"
)
```

```python
bar = alt.Chart(df).transform_filter(
    brush  # cross filter
).mark_bar().encode(
    x='league_type:N',
    y='count():Q',
    color='league_type:N',
    tooltip=[
        alt.Tooltip('league_type:N', title='League Type'),
        alt.Tooltip('count():Q', title='Count')
    ]
).properties(
    title="Bar Chart Filtered by Scatter Brush"
)
```

```python
scatter & bar
```

```python
bar_base = alt.Chart(df).mark_bar().encode(
    x='country:N',
    y='count():Q',
    color='country:N',
    tooltip=[
        alt.Tooltip('country:N', title='Country'),
        alt.Tooltip('count():Q', title='Count')
    ]
)

bar_text = bar_base.mark_text(
    dy=-5,
    color='black'
).encode(
    text='count():Q'
)

bar_layered = alt.layer(bar_base, bar_text).properties(
    title="Layered Bar Chart of Country Counts"
)
```
